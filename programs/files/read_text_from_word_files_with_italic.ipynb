```python
import docx
```

```python
file = './KTU 1.2_text processed.docx'
```

```python
word_doc = docx.Document(file)

for para in word_doc.paragraphs:
    print("WHOLE LINE: ", repr(para.text))
    
    for run in para.runs:
        if run.italic:
            print(repr(run.text), 'ITALIC')
        else:
            print(repr(run.text), 'NON-ITALIC')
            
    print()
```
